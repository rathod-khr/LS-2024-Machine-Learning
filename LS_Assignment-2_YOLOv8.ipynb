```python
# Install the necessary package
!pip install ultralytics

# Mount Google Drive
from google.colab import drive
import os
import yaml
from ultralytics import YOLO

# Mount Google Drive
#drive.mount('/content/drive')

# Define the base directory
#base_dir = '/content/drive/My Drive/path_to_dataset'  # Update with your actual dataset path
train_images_dir = 'train/images'
train_labels_dir = 'train/labels'
val_images_dir = 'valid/images'
val_labels_dir = 'valid/labels'
test_images_dir = 'test/images'
test_labels_dir = 'test/labels'

# Create a YAML file for the dataset configuration
dataset_config = {
    'train': train_images_dir,
    'val': val_images_dir,
    'test': test_images_dir,
    'nc': 2,  # Number of classes, update this
    'names': ['class1', 'class2']  # Replace with your actual class names
}

# Save the config file
with open('/content/dataset.yaml', 'w') as file:
    yaml.dump(dataset_config, file, default_flow_style=False)

# Load a YOLOv8 model, you can use 'yolov8n' for a small model or 'yolov8l' for a large model
model = YOLO('yolov8n.pt')

# Train the model
model.train(data='/content/dataset.yaml', epochs=20, imgsz=640)

# Evaluate the model on the validation set
metrics = model.val()
print("Validation Metrics:", metrics)

# Predict on the test dataset
results = model.predict(source=test_images_dir)

# Print results
for result in results:
    print(result)
```
